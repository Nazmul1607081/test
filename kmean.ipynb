```
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```
LABEL_COLOR_MAP = {0 : 'r',
                   1 : 'k',
                   2 : 'b',
                   3 : 'g'
                   }
```

```
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
label_color = [LABEL_COLOR_MAP[l] for l in y]
plt.scatter(X[:, 0], X[:, 1], 
            marker='.', s=200, linewidths=3, c=y)
```

```
kmeans = KMeans(n_clusters=3, init='random', max_iter=300, n_init=10, random_state=0)
kmeans.fit(X)
```

```
pred_y = kmeans.fit_predict(X)
label_color = [LABEL_COLOR_MAP[l] for l in pred_y]
```

```
plt.scatter(X[:, 0], X[:, 1])
plt.scatter(X[:, 0], X[:, 1], 
            marker='.', s=200, linewidths=3, c=pred_y)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
```
